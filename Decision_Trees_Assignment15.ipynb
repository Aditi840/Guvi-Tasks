**Importing the libraries**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

**Importing the dataset**

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
print(df)
print(df.columns)
print(df.isnull())
```

```python
df = pd.get_dummies(df, columns=['Gender'])
print(df)
```

**Splitting the dataset into the Training set and Test set**

```python
# Splitting the dataset into the Training set and Test set
X = df[['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']]
y = df['Purchased']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**Feature Scaling**

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Fitting Decision Tree Classification to the Training set**

```python
# Creating the decision tree classifier
classifier = DecisionTreeClassifier()

# Fitting the classifier to the training set
classifier.fit(X_train, y_train)
```

**Predicting the Test set results**

```python
# Making predictions on the test set
y_pred = classifier.predict(X_test)
print(y_pred)
```

**Making the Confusion Matrix**

```python
# Creating the confusion matrix
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

**Visualising the Training set results**

```python
print(X.shape)
print(y.shape)
# Select the variables of interest
X_train_plot = X_train[:, :2]
y_train_plot = y_train

# Create a color map
cmap = ListedColormap(['red', 'green'])

# Plot the data points
plt.scatter(X_train_plot[:, 0], X_train_plot[:, 1], c=y_train_plot, cmap=cmap)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Training Set')
plt.show()
```

**Visualising the Test set results**

```python
# Select the variables of interest
X_test_plot = X_test[:, :2]
y_test_plot = y_test

# Create a color map
cmap = ListedColormap(['red', 'green'])

# Plot the data points
plt.scatter(X_test_plot[:, 0], X_test_plot[:, 1], c=y_test_plot, cmap=cmap)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Test Set')
plt.show()
```
